```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import feature_selection_helpers as fsh
```

```python
df_topical = pd.read_sas('data/nsch_2019e_topical.sas7bdat')
```

```python
overlaps_list = []
for col in df_topical.select_dtypes(include=['number']):
    #print(col)
    overlaps_list.append(fsh.hist_overlap(col,df_topical))
```

```python
# Make dataframe of the overlap scores
overlaps_df = pd.DataFrame(overlaps_list, index=list(df_topical.select_dtypes(include=['number'])))
```

```python
# Sort and print the features with the 40 lowest overlap scores.
# Note that K7Q02R_R is the variable we're conditioning on, so the two histograms are entirely different (have different ranges)
# FWC is a sampling weight and doesn't really provide any usable information
overlaps_df.sort_values(by=0).iloc[:40]
```
